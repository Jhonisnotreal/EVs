# Importing Libraries

```python
import pandas as pd
import numpy as np
```

# Reading the Data

```python
data = pd.read_csv('Dataset/Electric_Vehicle_Population_Data.csv')
```

```python
data
```

## Replacing Data

```python
data['Vehicle Location'] = data['Vehicle Location'].str.replace('POINT (', '')
```

```python
data['Vehicle Location']
```

```python
data['Vehicle Location'] = data['Vehicle Location'].str.replace(')', '')
```

## Cleaning Null Values

```python
data.isnull().sum().sum()
```

```python
(((data.isnull().sum() / len(data)) * 100)).sort_values(ascending=False)
```

```python
data[['Legislative District', 'Vehicle Location', '2020 Census Tract', 'City', 'Electric Utility', 'Postal Code', 'County']]
```

```python
data['Legislative District'] = data['Legislative District'].fillna(np.mean(data['Legislative District']), inplace=True)
```

```python
data.fillna("No data", inplace=True)
```

## No more null values

```python
data.isnull().sum().sum()
```

## Separating geographic coordenates in it's respective columns

```python
data[['X', 'Y']] = data['Vehicle Location'].str.split(' ', n=1, expand=True)
```

```python
data[['X', 'Y']]
```

## Deleting the original column

```python
data.drop(['Vehicle Location'], axis=1, inplace=True)
```

## Last look

```python
print(data.columns)
print(data.shape)
```

## Saving clean data

```python
data.to_csv('Dataset/data_clean.csv', index=False)
```
